```python
from IPython.display import clear_output
import time
import sys
```

```python
class CCROI:
    
    def __init__(self,prop):
        self.prop = prop
        self.inex = []
        self.inex2 = []
        self.cashflow = []
        self.inex3 = []
        self.acf = []
        self.ccr = []
        
#    BOX1/ inex
    def income(self):
        self.inex = 0
        
        
        print("\t-- Howdy Partner --\n -- Let's calculate your Income --")
        time.sleep(1)
        try:
            income = int(input(f'What is your rental income? '))
            time.sleep(.2)
            laundry = int(input(f'How much is your laundry? '))
            time.sleep(.2)
            stroage = int(input(f'How much is your stroage? '))
            time.sleep(.2)
            misc = int(input(f'How much is your misc? '))
            time.sleep(.2)
            clear_output()
            
        except ValueError:
            print(f"Not an integer!")
            return ccr.income()
            clear_output()
        else:
            self.inex += income
            self.inex += laundry
            self.inex += stroage
            self.inex += misc
#             print(f"-- your number was added --\n")
            
#             time.sleep(2)
            print(f"    -- calculating --")
            time.sleep(1)
            for i in range(10):
                sys.stdout.write(" .")
                sys.stdout.flush()
                time.sleep(.2)
            clear_output()
        print("----------------------------")    
        print(f'Total Montly Income: ${self.inex}')
        print("----------------------------\n")
        time.sleep(2)
        
            

            
            
            
            
    #    BOX2/ Inex2
    
    def expences(self):
#         expences_total = 0
        self.inex2 = 0
        utilities = 0
        print(f"-- Now let's calculate your Expences --\n")
      
        try:
            u = input(f'If you know your Utility Total press(Y) If not (N).')
            if u == "n":
                electric = int(input(f'How much is your electric?'))
                gas = int(input(f'How much is your Gas?'))
                water = int(input(f'How much is your Water?'))
                sewer = int(input(f'How much is your Sewer?'))
                utilities += electric
                utilities += gas
                utilities += water
                utilities += sewer
                self.inex2 += utilities
                print("\nYour Utilities were added\nPlease continue.")
                time.sleep(.2)
                clear_output()
                
            elif u == 'y':
                
                utility = int(input(f'How much are your Utilities: ')) 
#                 expences_total += utility
                self.inex2 += utility
                time.sleep(.2)
                clear_output()

            tax = int(input(f'How much is your Taxes? '))
            insurance = int(input(f'How much is your Insurance? '))
            hoa = int(input(f'How much is your HOA?'))
            lawn_snow = int(input(f'How much is your Lawn & Snow removal? '))
            vacancy = int(input(f'How much is your Vacancy? '))
#             repairs = int(input(f'How much is your Repairs?'))
#             capex = int(input(f'How much is your CapEx?'))
#             prop_manage = int(input(f'How much is your Property Management?'))
#             mortagage = int(input(f'How much is your mortagage?'))
            time.sleep(.2)
            clear_output()
        
        except ValueError:
            print("Not an integer!")
            return ccr.expences()
            clear_output()
        else:
            
            
            self.inex2 += utilities
            self.inex2 += tax
            self.inex2 += insurance
            self.inex2 += hoa
            self.inex2 += lawn_snow
            self.inex2 += vacancy
            
        print("your Expense number has been added.")
        print(f"    -- calculating --")
        time.sleep(1)
        for i in range(10):
            sys.stdout.write(" .")
            sys.stdout.flush()
            time.sleep(.2)
        clear_output()
        
        print("----------------------------") 
        print(f'Total Montly Expenses: ${self.inex2}')
        print("----------------------------\n")     
        
        clear_output()

        
        
        
        
    # BOX3 box1-box2 = Total Montly Cash Flow = self.cashflow
    
    def in_ex(self):
        inex = self.inex
        inex2 = self.inex2
        self.cashflow = 0
        total = inex - inex2
        self.cashflow += total
#         if inex > inex2:
#             total = inex - inex2
#         else:
#             total = inex2 - inex
        
        
        print(f'Total Montly Income: ${self.inex}')
        print("----------------")
        time.sleep(2)
        print(f'Total Montly Expences: ${self.inex2}') 
        print("----------------")
        time.sleep(1)
        print(f"Total Monthly Cash Flow: ${self.cashflow}")
        print("----------------")
        time.sleep(2)
    

    
    
    
    
    #  BOX4/inex3 investment
    
    def CCR(self):
        self.inex3 = 0
        
        print("\n-- let's calculate your Investment --\n")
      
        try:
            down_payment = int(input(f'How much is your down_payment? '))
            closing_cost = int(input(f'How much is your closing_cost? '))
            rehab = int(input(f'How much is your rehab? '))
            misc = int(input(f'How much is your misc? '))
            
        except ValueError:
            print("Not an integer!")
            return ccr.CCR()
            clear_output()
        else:
            self.inex3 += down_payment
            self.inex3 += closing_cost
            self.inex3 += rehab
            self.inex3 += misc
            print("\nyour Total Investment number was added.\n")
            

        clear_output()
        time.sleep(3)    
        print(f'Total Investment: ${self.inex3}')
       
        
    
#     Box 5 Annual Cash flow
    def ac_flow(self):
        self.ccr = float(0)
        acf = self.cashflow * 12
        self.ccr += acf/self.inex3 *100
        
        for i in range(10):
            sys.stdout.write(" .")
            sys.stdout.flush()
            time.sleep(.1)
        clear_output()
        print("-----------------------------------")
        print(f'\tMontly Income: ${self.inex}')
        print("-----------------------------------")
        time.sleep(2)
        print(f'\tMontly Expences: ${self.inex2}') 
        print("-----------------------------------")
        time.sleep(1)
        print(f"\tMonthly Cash Flow: ${self.cashflow}")
        print("-----------------------------------")
        print(f'\tInvestment: ${self.inex3}')
        print("-----------------------------------")
        print(f'\t-- Good News --')
        time.sleep(2)
        print("-----------------------------------")
        print(f'\tROI: %{self.ccr}')
        print("-----------------------------------")
        print("\t-- See you next time --")
        
        

        
        
        
        
        
        

ccr = CCROI(2000)


ccr.income()
ccr.expences()
ccr.in_ex()
ccr.CCR()
ccr.ac_flow()
# ccr.ccr_percent()

# print("----------------")
# ccr.expences()
# print("----------------")
# ccr.CCR()
# print("----------------")
# print(ccr.inex)
# print("----------------")
# print(ccr.inex2)
```
